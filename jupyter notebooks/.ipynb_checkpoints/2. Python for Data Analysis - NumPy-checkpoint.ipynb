#**Numpy indexing and selection**

```python
a=np.arange(0,11)
a
```

```python
a[8]
```

```python
a[1:5]
```

```python
a[:6]
```

```python
a[2:]
```

```python
#Setting a value with index range (Broadcasting)
a[0:3]=15
a
```

```python
# Reset array
a = np.arange(0,11)
#Show
a
```

```python
# Slices
slice_of_a = a[0:6]
#Show slice
slice_of_a
```

```python
#Change Slice
slice_of_a[:]=99
#Show Slice again
slice_of_a
```

```python
#Now note the changes also occur in our original array!
a
```

```python
#Data is not copied, it's a view of the original array! This avoids memory problems!
#To get a copy, need to be explicit
a_copy = a.copy()
a_copy
```

```python
a_copy[:]=100
a_copy
```

```python
a
```
