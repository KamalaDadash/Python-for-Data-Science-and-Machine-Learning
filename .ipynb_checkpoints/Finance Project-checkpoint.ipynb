## Data

We need to get data using pandas datareader. We will get stock information for the following banks:
*  Bank of America
* CitiGroup
* Goldman Sachs
* JPMorgan Chase
* Morgan Stanley
* Wells Fargo

** Our aim is to get the stock data from Jan 1st 2006 to Jan 1st 2016 for each of these banks. We set each bank to be a separate dataframe, with the variable name for that bank being its ticker symbol. This will involve a few steps:**
1. Use datetime to set start and end datetime objects.
2. Figure out the ticker symbol for each bank.
2. Figure out how to use datareader to grab info on the stock.


```python
from datetime import datetime
start = datetime(2006, 1, 1)
end=datetime(2016,1,1)
```

```python
# Bank of America
BAC = data.DataReader("BAC", 'yahoo', start, end)

# CitiGroup
C = data.DataReader("C", 'yahoo', start, end)

# Goldman Sachs
GS = data.DataReader("GS", 'yahoo', start, end)

# JPMorgan Chase
JPM = data.DataReader("JPM", 'yahoo', start, end)

# Morgan Stanley
MS = data.DataReader("MS", 'yahoo', start, end)

# Wells Fargo
WFC = data.DataReader("WFC", 'yahoo', start, end)
```

```python
BAC.head()
```

** We create a list of the ticker symbols (as strings) in alphabetical order.Let's call this list: tickers**

```python
tickers=["BAC","C","GS","JPM","MS","WFC"]
```

** We use pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks ande set the keys argument equal to the tickers list. Also pay attention to what axis you concatenate on.**

```python
bank_stocks=pd.concat([BAC,C,GS,JPM,MS,WFC],keys=tickers,axis=1)
bank_stocks.head()
```

```python
type(BAC)
```

** We set the column name levels:**

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

** We check the head of the bank_stocks dataframe.**

```python
bank_stocks.head()
```

# EDA

Before continuing, check out the documentation on [Multi-Level Indexing](http://pandas.pydata.org/pandas-docs/stable/advanced.html) and [Using .xs](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.xs.html).


** What is the max Close price for each bank's stock throughout the time period?**

```python
bank_stocks.xs('Close',level='Stock Info',axis=1).max()# easy way. it is automatically axis=0 mode.
```

```python
bank_stocks.xs('Close',level='Stock Info',axis=1)
```

** Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns=pd.DataFrame()
```

** We can use pandas pct_change() method on the Close column to create a column representing this return value. 

```python
tickers=["BAC","C","GS","JPM","MS","WFC"]
for i in tickers:
    returns[i+' Return']=bank_stocks[i]['Close'].pct_change()
returns.head()  
```

** We create a pairplot using seaborn of the returns dataframe.

```python
import seaborn as sns
sns.pairplot(returns[1:])
sns.set_style('darkgrid')
```

** Using this returns DataFrame, figure out on what dates each bank stock had the best and worst single day returns. **

```python
returns.idxmin()
```

** You should have noticed that Citigroup's largest drop and biggest gain were very close to one another, did anythign significant happen in that time frame? **

```python
returns.idxmax()
```

** We take a look at the standard deviation of the returns, which stock would you classify as the riskiest over the entire time period? Which would you classify as the riskiest for the year 2015?**

```python
returns.std()
```

```python
returns.loc['2015'].std()
```

** We create a distplot using seaborn of the 2015 returns for Morgan Stanley **

```python
sns.displot(returns['MS Return'].loc['2015'],color='green',bins=50,kde=True)
sns.set_style('whitegrid')
```

** We create a distplot using seaborn of the 2008 returns for CitiGroup **

```python
sns.displot(returns['C Return'].loc['2008'],color='red',bins=100,kde=True)
sns.set_style('whitegrid')
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

** We plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
plt.figure(figsize=(12,6))
plt.plot(bank_stocks['BAC']['Close'].loc['2008'],label='BAC Close')
plt.plot(bank_stocks['BAC']['Close'].rolling(window=30).mean().loc['2008'],label='30 day avg')
plt.legend()
```

** We create a heatmap of the correlation between the stocks Close Price.**

```python
tc=bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr()
sns.heatmap(tc,annot=True)
```

** Optional: Use seaborn's clustermap to cluster the correlations together:**

```python
sns.clustermap(tc)
```

```python
tc.iplot(kind='heatmap',colorscale='rdylbu')
```
